```python
import random
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Step 1: Generate a 6-digit OTP
def generate_otp():
    otp = ''.join([str(random.randint(0, 9)) for _ in range(6)])
    return otp

# Step 2: Send OTP via Email
def send_otp_via_email(receiver_email, otp):
    sender_email = "artikumariabps96@gmail.com"  # Replace with email
    sender_password = "nsjq jspa mybv onyo"      # Replace with app password

    try:
        # Create the email content
        message = MIMEMultipart()
        message['From'] = sender_email
        message['To'] = receiver_email
        message['Subject'] = "Your OTP Verification Code"
        body = f"Your OTP is: {otp}"
        message.attach(MIMEText(body, 'plain'))

        # Connect to the SMTP server and send the email
        with smtplib.SMTP('smtp.gmail.com', 587) as server:
            server.starttls()
            server.login(sender_email, sender_password)
            server.sendmail(sender_email, receiver_email, message.as_string())

        print("OTP sent successfully!")
    except Exception as e:
        print("Error sending OTP:", e)

# Step 3: Validate OTP
def validate_otp(expected_otp):
    attempts = 3  # Allow 3 attempts
    while attempts > 0:
        entered_otp = input("Enter the OTP: ")
        if entered_otp == expected_otp:
            print("Access Granted!")
            return True
        else:
            attempts -= 1
            print(f"Incorrect OTP. You have {attempts} attempts left.")

    print("Access Denied!")
    return False

# Step 4: OTP Verification System
def otp_verification_system():
    receiver_email = input("Enter your email address: ")
    otp = generate_otp()
    send_otp_via_email(receiver_email, otp)
    validate_otp(otp)

# main_function
otp_verification_system()
```
